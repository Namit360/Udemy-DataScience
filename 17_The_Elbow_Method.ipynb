```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

```python
data = pd.read_csv('Categorical.csv')
data
```

## Map the data

```python
data_mapped = data.copy()
data_mapped['continent'] = data_mapped['continent'].map({'North America':0,'Europe':1,'Asia':2,'Africa':3,'South America':4, 'Oceania':5,'Seven seas (open ocean)':6, 'Antarctica':7})
data_mapped
```

## Select the features

```python
x = data_mapped.iloc[:,3:4]
x
```

## Clustering

```python
kmeans = KMeans(7) # 7 indicates no of clusters
```

```python
kmeans.fit(x)
```

## Clustering Results

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
data_cluster = data.copy()
data_cluster['Clusters'] = identified_clusters
data_cluster
```

## Plot the data

```python
plt.scatter(data['Longitude'], data['Latitude'],c=data_cluster['Clusters'], cmap = 'rainbow')
plt.xlim(-180,180)
plt.ylim(-90, 90)
plt.show()
```

##Selecting the no.of Clusters

###WCSS

```python
kmeans.inertia_
```

```python
wcss=[]

for i in range(1,11):
  kmeans = KMeans(i)
  kmeans.fit(x)
  wcss_iter = kmeans.inertia_
  wcss.append(wcss_iter)
```

```python
wcss
```

###The Elbow Method

```python
n_clusters = range(1,11)
plt.plot(n_clusters,wcss)
plt.title('The Elbow Method')
plt.xlabel('No of Cluster')
plt.ylabel('Within Cluster Sun of Squares')
```
